# System Dynamics Modelling with Python

### Importing necessary packages

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Module 1: Unconstrained Growth and Decay

Equation: population(i+1) = population(i) + growth rate x population(i) x delta time

Description: Use Malthusian Growth Model to stimulate the growth of bacteria population for 48 hours with initial population of 100 and growth rate of 0.1.

### Finite Difference Method

```python
# initialising variables
sim_length = 48
growth_rate = 0.1
delta_t = 0.01
num_iter = int(sim_length/delta_t)
population = np.zeros(num_iter)
population[0] = 100
```

```python
for i in range(num_iter-1):
  growth = growth_rate * population[i]
  population[i+1] = population[i] + growth * delta_t
```

```python
x_axis = np.arange(0, sim_length, delta_t)
```

```python
plt.plot(x_axis, population)
plt.xlabel('Time (Hour)')
plt.ylabel('Population')
plt.title('Malthusian Growth Model for Bacteria Population Growth')
plt.show()
```

### Analytical Solutions Method

```python
def P(t):
  return 100*np.exp(growth_rate*t)
```

```python
time = np.arange(0, sim_length, delta_t)
pop = P(time)
```

```python
plt.plot(time, pop)
plt.xlabel('Time (Hour)')
plt.ylabel('Population')
plt.title('Malthusian Growth Model for Bacteria Population Growth')
plt.show()
```

### Finite Difference Method versus Analytical Method

```python
print("Finite Difference Method final population:", population[-1])
print("Analytical Dolutions Method final population:", pop[-1])
```

The difference comes only when looking at a small-number simulation. The error comes from the discretisation of the Finite Difference Method. The error could be minimised by reducing the delta t variable. But it will slow down the program due to large number of iterations.
